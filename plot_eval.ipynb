```python
import numpy as np
import matplotlib.pyplot as plt
import dill
from statsmodels.distributions.empirical_distribution import ECDF

%matplotlib inline  
```

```python
results=dill.load( open( "eval/Stoch_Hybrid_ieee300_100iter_3pwl_gurobi_mingap_cluster_det_conc_warm_start.p", "rb" ) )
```

```python
results.keys()
```

```python
np.mean(results['Q_dict'][str(0)])
```

```python
plt.hist(results['Q_dict'][str(0)],bins=100)
plt.show()
```

```python
ecdf1=ECDF(results['Q_dict'][str(0)])
ecdf2=ECDF(results['Q_dict'][str(-1)])
plt.plot(ecdf1.x,ecdf1.y)
plt.plot(ecdf2.x,ecdf2.y)
plt.show()
```

```python
for k in reversed(range(0,100,5)+[-1]):
    ecdf=ECDF(results['Q_dict'][str(k)])
    if k ==-1:
        k=0
    colors=str((100-k)/200.+0.1)
    plt.plot(ecdf.x,ecdf.y,color=colors)
plt.xlim(4500,5800)
plt.show()
```

```python
Q_0=np.mean(results['Q_dict'][str(0)])
```

```python
plt.plot([(np.percentile(results['Q_dict'][str(k)],95)-Q_0)/Q_0*100 for k in range(0,100,5)+[-1] ])
plt.plot([(np.percentile(results['Q_dict'][str(k)],5)-Q_0)/Q_0*100 for k in range(0,100,5)+[-1] ])
plt.plot([ (np.mean(results['Q_dict'][str(k)])-Q_0)/Q_0*100 for k in range(0,100,5)+[-1] ])
plt.show()
```
